**7. Display age statistics for each race (feature race) and each gender. Use groupby and describe. In this way, find the maximum age of men of the Asian-Pac-Islander race.**

**8. Among whom is the proportion of high earners (>50K) greater: married or single men (feature marital-status)? Consider as married those whose marital-status starts with Married (Married-civ-spouse, Married-spouse-absent, or Married-AF-spouse); consider the rest as single.**

```python
# your code
men_data = data[data['sex'] == 'Male'].copy()

def set_status(row):
    if row['marital-status'] in ('Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'):
        return 'Married'
    else:
        return 'Single'

men_data['married_status'] = men_data.apply(set_status, axis='columns')
proportion = men_data.groupby('married_status')['salary'].value_counts(normalize=True)

proportion
```

**9. What is the maximum number of hours a person works per week (feature hours-per-week)? How many people work this number of hours, and what is the percentage of high earners among them?**

```python
# your code
max_load = data['hours-per-week'].max()
print(f"Max hour: {max_load} hour/week")

num_workaholics = data[data['hours-per-week'] == max_load].shape[0]
print(f"Number of people who work max hours per week: {num_workaholics} kishi")

rich_share = (data[data['hours-per-week'] == max_load]['salary'] == '>50K').mean()

print(f"Proportion: {rich_share:.2%}")
```

**11.Group people into age groups: young, adult, and retiree, where:**
* *young* corresponds to 18-35 years old
* *adult* - 35-70 years old
* *retiree* - 70-100 years old

**Assign the name of the corresponding group to each person in a new column called 'AgeGroup'.**

```python
# your code
def set_age(row):

    if row.age >= 17 and row.age <= 35:
        return 'young'
    elif row.age > 35 and row.age <= 70:
        return 'adult'
    else:
        return 'retiree'
    
data['AgeGroup'] = data.apply(set_age, axis='columns')

data
```

**12-13. "Determine the number of people earning >50K in each of the age groups (column 'AgeGroup'), and also output the name of the age group where people more frequently earn more than 50K (>50K).**

```python
# your code
rich_people = data[data['salary'] == '>50K']

rich_counts_by_age = rich_people.groupby('AgeGroup').size()

print(rich_counts_by_age)

most_rich_group = rich_counts_by_age.idxmax()

print(f"\nThe most rich group is {most_rich_group}s")
```

**14. Group people by occupation (column 'occupation') and determine the number of people in each group. After that, write a filtering function 'filter_func' that will return only those groups where the average age (column 'age') is no more than 40 and all workers work more than 5 hours per week (column 'hours-per-week').**

```python
occupation_counts = data.groupby('occupation').size()

print("Number of people in each occupation group:")
print(occupation_counts)

def filter_func(group):
    avg_age_check = group['age'].mean() <= 40
    min_hours_check = (group['hours-per-week'] > 5).all()

    return avg_age_check and min_hours_check

filtered_data = data.groupby('occupation').filter(filter_func)

remaining_occupations = filtered_data['occupation'].unique()

print("\nOccupations that satisfy the conditions:")
print(remaining_occupations)
```
